### Use pd.cut or pd.qcut to create 4 groups (bins) of cars, by Year. We want to explore how cars have evolved decade by decade

```python
cars['Year'].value_counts()
```

```python
def classifier(x):
    if x in range(1980,1990):
        return "80's"
    elif x in range(1990,2000):
        return "90's"
    elif x in range(2000,2010):
        return "10's"
    else:
        return "20's"
```

```python
cars["Decade"]= cars.Year.apply(lambda x: classifier(x))
cars[['Year','Decade']]
```

```python
cars["Year_range"]= pd.cut(cars["Year"], 
                             bins = [1980,1989,1999,2009,2019],
                             labels=["80s", "90s", "00s", "10s"])

cars.loc[:,['Year','Decade','Year_range']]
```
