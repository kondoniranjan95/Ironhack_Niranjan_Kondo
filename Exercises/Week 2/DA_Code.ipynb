```python
import pandas as pd
import numpy as np
```

### extracting data from the files

```python
file1 = pd.read_csv('file1.csv')
file2= pd.read_csv('file2.csv')
file3= pd.read_csv('file3.csv')
```

### Working on the data

```python
main_data=pd.concat([file1,file2,file3],axis=0)
main_data
```

### Renaming the headers of the data

```python
main_data.columns= [item.lower() for item in main_data.columns]

main_data
```

### deleting the column "customer"

```python
main_data= main_data.rename(columns={'st':'state',"customer lifetime value":"cust_life_val",
                          "monthly premium auto":"month_prem_auto",
                          "number of open complaints":"no_open_compl",
                          "policy type":"policy_type","vehicle class":"vehicle_class",
                         "total claim amount":"total_claim_amt"})
```

```python
main_data= main_data.drop(columns=["customer"])
```

```python
main_data.info()
```

### customer life value data cleaning

```python
main_data["cust_life_val"]= main_data.cust_life_val.replace("%","")
```

```python
main_data["cust_life_val"]= pd.to_numeric(main_data["cust_life_val"],errors='coerce')
```
