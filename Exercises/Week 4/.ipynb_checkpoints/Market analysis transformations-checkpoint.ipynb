### One Hot/Label Encoding (categorical).

```python
main_data["state"].value_counts()
state_types = ('california','oregon ','arizona','nevada','washington','Unknown')
state_df = pd.DataFrame(state_types, columns=['state_types'])

# converting type of columns to 'category'
state_df['state_types'] = state_df['state_types'].astype('category')

# Assigning numerical values and storing in another column
state_df['state_types_cat'] = state_df['state_types'].cat.codes
state_df
```

```python
from sklearn.preprocessing import OneHotEncoder

# creating instance of one-hot-encoder
enc = OneHotEncoder(handle_unknown='ignore')

# passing state-types-cat column (label encoded values of state_types)
enc_df = pd.DataFrame(enc.fit_transform(state_df).toarray())

# merge with main df bridge_df on key values
state_df = state_df.join(enc_df)
state_df
```
